```python
import random
import pickle

# Define the game board
def display_board(board):
    print()
    print(board[0] + ' | ' + board[1] + ' | ' + board[2])
    print('--+---+--')
    print(board[3] + ' | ' + board[4] + ' | ' + board[5])
    print('--+---+--')
    print(board[6] + ' | ' + board[7] + ' | ' + board[8])
    print()

# Check if a player has won
def check_winner(board, player):
    win_positions = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),
        (0, 3, 6), (1, 4, 7), (2, 5, 8),
        (0, 4, 8), (2, 4, 6)
    ]
    for pos in win_positions:
        if board[pos[0]] == board[pos[1]] == board[pos[2]] == player:
            return True
    return False

# Check if the game is a tie
def is_tie(board):
    return ' ' not in board

# AI using Q-learning
class TicTacToeAI:
    def __init__(self, symbol, alpha=0.5, gamma=0.9, epsilon=0.2):
        self.symbol = symbol
        self.q_table = {}
        self.alpha = alpha    # learning rate
        self.gamma = gamma    # discount factor
        self.epsilon = epsilon  # exploration rate
        self.last_state = None
        self.last_action = None

    def get_state(self, board):
        return ''.join(board)

    def choose_action(self, board):
        state = self.get_state(board)
        actions = [i for i in range(9) if board[i] == ' ']

        # Exploration vs Exploitation
        if random.uniform(0, 1) < self.epsilon:
            action = random.choice(actions)
        else:
            q_values = [self.q_table.get((state, a), 0) for a in actions]
            max_q = max(q_values)
            best_actions = [a for a, q in zip(actions, q_values) if q == max_q]
            action = random.choice(best_actions)

        # Store last move
        self.last_state = state
        self.last_action = action

        return action

    def update_q_value(self, reward, new_board):
        if self.last_state is None:
            return

        old_q = self.q_table.get((self.last_state, self.last_action), 0)
        new_state = self.get_state(new_board)
        future_qs = [self.q_table.get((new_state, a), 0) for a in range(9) if new_board[a] == ' ']

        if future_qs:
            max_future_q = max(future_qs)
        else:
            max_future_q = 0

        # Q-learning formula
        new_q = old_q + self.alpha * (reward + self.gamma * max_future_q - old_q)
        self.q_table[(self.last_state, self.last_action)] = new_q

# Main game loop
def play_game(ai_player=None):
    board = [' ' for _ in range(9)]
    current_player = 'X'

    while True:
        display_board(board)

        if current_player == 'X':
            try:
                move = int(input("Choose a position (1-9): ")) - 1
                if move < 0 or move > 8 or board[move] != ' ':
                    print("Invalid move. Try again.")
                    continue
            except ValueError:
                print("Please enter a number between 1 and 9.")
                continue
        else:
            move = ai_player.choose_action(board)
            print(f"Computer chooses position {move + 1}")

        board[move] = current_player

        if check_winner(board, current_player):
            display_board(board)
            if current_player == 'X':
                print("Congratulations! You win! ðŸŽ‰")
                ai_player.update_q_value(-1, board)
            else:
                print("Computer wins! ðŸ’»")
                ai_player.update_q_value(1, board)
            break

        if is_tie(board):
            display_board(board)
            print("It's a tie!")
            ai_player.update_q_value(0.5, board)
            break

        # Reward for each move
        if current_player == 'O':
            ai_player.update_q_value(0, board)

        current_player = 'O' if current_player == 'X' else 'X'

# Main program
if __name__ == '__main__':
    ai = TicTacToeAI('O')

    while True:
        play_game(ai)
        choice = input("Play again? (yes/no): ").lower()
        if choice != 'yes':
            break

    # Optional: Save Q-table for future use
    with open('tic_tac_toe_qtable.pkl', 'wb') as f:
        pickle.dump(ai.q_table, f)

    print("Thanks for playing!")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# Read the data
data = pd.read_csv('company_sales_data.csv')

# Extract required columns
months = data['month_number']
total_profit = data['total_profit']

# Create line plot
plt.plot(months, total_profit, marker='o', color='b', linewidth=2)
plt.title('Company Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.xticks(months)
plt.yticks(range(100000, 500001, 50000))
plt.show()
```

The chart shows that the company experienced fluctuations in profit throughout the year, with a noticeable increase between August and November. The highest profit was recorded in November, while February saw the lowest. This suggests stronger sales performance towards the yearâ€™s end, possibly due to seasonal demand or promotions.

```python
# Extract required columns
bathingsoap_sales = data['bathingsoap']
facewash_sales = data['facewash']

# Create subplots
plt.figure(figsize=(10, 6))

# Subplot for Bathing soap
plt.subplot(2, 1, 1)
plt.plot(months, bathingsoap_sales, marker='o', color='g')
plt.title('Sales Data for Bathing Soap')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.grid(True)

# Subplot for Facewash
plt.subplot(2, 1, 2)
plt.plot(months, facewash_sales, marker='o', color='m')
plt.title('Sales Data for Facewash')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.grid(True)

# Adjust layout and show
plt.tight_layout()
plt.show()
```

The charts display monthly sales trends for Bathing Soap and Facewash. Both products maintained relatively steady sales throughout the year, with minor fluctuations. Bathing Soap consistently recorded higher sales compared to Facewash. These trends suggest stable consumer demand, with potential seasonal influences in specific months.
